# 1. 从fundemantals.csv开始！

fundemantals.csv 是这些股票的年报数据

请用数据回答以下问题：

1. S&P500股票在2015年`net income`的均值是多少？最大值比最小值多多少？（每问10分，共计20分）
2. S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是多少？固定资产占总资产比例最小的股票是的代码（ticker symbol）是什么？（每问10分，共计20分）


```python
fund_df = pd.read_csv('fundamentals.csv') #导入数据
data=fund_df.groupby(fund_df['For Year'].apply(lambda x:x if x==2015 else None)) #取年份为2015的数据
print('1.S&P500股票在2015年net income的均值是:')
data['Net Income'].mean()
```

```python
print('最大值比最小值多：')
data['Net Income'].max()-data['Net Income'].min()
```

```python
data=fund_df.groupby(fund_df['For Year'].apply(lambda x :x if x==2016 else None)) #取年份为2016的数据
print('2.S&P500股票在2016年的固定资产（fixed assets）占总资产(total assets)比例的均值是:')
data.apply(lambda x:fund_df['Fixed Assets']/fund_df['Total Assets']).mean(axis=1)
```

```python
t=data.apply(lambda x:fund_df['Fixed Assets']/fund_df['Total Assets']).min(axis=1) #固定资产占总资产比例最小的值
print('最小值为:\n'+str(t)+'\n')
result=fund_df[(fund_df['Fixed Assets']/fund_df['Total Assets']).isin(t)]['Ticker Symbol'] #固定资产占总资产比例为最小值的股票的代码
print('固定资产占总资产比例最小的股票共有{}个,代码分别为：'.format(result.agg("count")))
result
```

securities.csv包含了这些股票的基本信息

```python
#分组后取'ttt‘列’
data=sec_df_.groupby(['GICS Sector','Security'])[['ttt']].max()
print('各个sector中的加入时间最早的股票名称分别为:')
#取ttt的最小值，即加入时间最早的股票名称，将日期为0的数据替换为1000000，避免影响寻找最小值
data.dropna(0).groupby(['GICS Sector'])[['ttt']].apply(lambda x:x.replace(0,1000000).sort_values(by='ttt')[:1])
```

```python
#分组后取'ttt‘列’
data=sec_df_.groupby(['Address of Headquarters','Security'])[['ttt']].max()
print('每一个州中加入时间最晚的股票名称分别为:')
#取ttt的最大值，即加入时间最晚的股票名称
data.groupby(['Address of Headquarters'])[['ttt']].apply(lambda x:x.sort_values(by='ttt')[-1:])
```
