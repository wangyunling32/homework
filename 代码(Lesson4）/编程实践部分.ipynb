4. 打乱数据集

```python
x = ['快递太慢了！','不好吃','特别难吃','要齁死我了','很划算','下次还来','味道很不错！','香']
y = ['差评','差评','差评','差评','好评','好评','好评','好评']
x=np.array(x)
y=np.array(y)
def shuffle(x,y):
    permutation = np.random.permutation(y.shape[0])
    x = x[permutation]
    y = y[permutation]
    return x,y

# print result for certify
x,y=shuffle(x,y)
for i,j in zip(x,y):
    print(i,':',j)
```
