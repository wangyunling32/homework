```python
import jieba
import sys
vocab=[]
word2id={}
char_x2=input('请输入文本:\n')
test2 = jieba.cut(char_x2, cut_all=False)
text= " ".join(test2)
vocab=list(set(text.split(' ')))
for i in range(len(vocab)):
    word2id[vocab[i]]=i
id2word={value:key for key,value in word2id.items()}
print('\nvocab = '+str(vocab))
print('\nword2id = '+str(word2id))
print('\nid2word = '+str(id2word))
```
